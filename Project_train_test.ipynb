# Preparing the Data

```python
from sahi.utils.coco import Coco, CocoCategory, CocoImage, CocoAnnotation
from sahi.utils.file import save_json, load_json
from sahi.slicing import slice_coco
import PIL as pil
import numpy as np
import random
import cv2
import pandas as pd
from tqdm import tqdm
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

## Create Coco style annotation files

## upsampled-2x

```python
def save_dataset_json(img_path, label_path, output_path, categories, category_id_counter):
    coco = Coco()
    for k,v in categories.items():
        coco.add_category(CocoCategory(id=v, name=k))
    for img_name in tqdm(os.listdir(img_path)):
        with pil.Image.open(os.path.join(img_path, img_name)) as im:
            img_arr = np.asarray(im)
            try:
                height, width, _ = img_arr.shape
            except ValueError:
                img_arr = np.expand_dims(img_arr, axis=-1)
                height, width, _ = img_arr.shape
                rgb_im = cv2.cvtColor(img_arr, cv2.COLOR_GRAY2RGB)
                pil.Image.fromarray(rgb_im).save(os.path.join(img_path, img_name))
        coco_image = CocoImage(file_name=img_name, height=height, width=width)
        img_label_path = os.path.join(label_path, img_name.split('.')[0] + '.txt')
        df = pd.read_csv(
            img_label_path, 
            sep=' ', 
            names=['x1', 'y1', 'x2', 'y2', 'x3', 'y3', 'x4', 'y4', 'class_name', 'difficult']
        )
        df['xmin'] = 2*(df[['x1', 'x2', 'x3', 'x4']].min(axis=1))
        df['xmax'] = 2*(df[['x1', 'x2', 'x3', 'x4']].max(axis=1))
        df['ymin'] = 2*(df[['y1', 'y2', 'y3', 'y4']].min(axis=1))
        df['ymax'] = 2*(df[['y1', 'y2', 'y3', 'y4']].max(axis=1))
        df['width'] = df['xmax'] - df['xmin']
        df['height'] = df['ymax'] - df['ymin']
        df = df[['xmin', 'ymin', 'width', 'height', 'class_name']]

        for index, row in df.iterrows():
            category_name = row['class_name']
            try: 
                category_id = categories[category_name]
            except KeyError:
                categories[category_name] = category_id_counter
                category_id = category_id_counter
                category_id_counter += 1
                coco.add_category(CocoCategory(id=category_id, name=category_name))

            coco_image.add_annotation(
                CocoAnnotation(
                      bbox=[
                        row['xmin'], 
                        row['ymin'], 
                        row['width'], 
                        row['height']
                      ],
                      category_id=category_id,
                      category_name=category_name
                    )
                )

        coco.add_image(coco_image)
    coco_json = coco.json      
    save_json(coco_json, output_path)
    return categories, category_id_counter
```

## upsampled-1x

```python
def save_dataset_json(img_path, label_path, output_path, categories, category_id_counter):
    coco = Coco()
    for k,v in categories.items():
        coco.add_category(CocoCategory(id=v, name=k))
    for img_name in tqdm(os.listdir(img_path)):
        with pil.Image.open(os.path.join(img_path, img_name)) as im:
            img_arr = np.asarray(im)
            try:
                height, width, _ = img_arr.shape
            except ValueError:
                img_arr = np.expand_dims(img_arr, axis=-1)
                height, width, _ = img_arr.shape
                rgb_im = cv2.cvtColor(img_arr, cv2.COLOR_GRAY2RGB)
                pil.Image.fromarray(rgb_im).save(os.path.join(img_path, img_name))
        coco_image = CocoImage(file_name=img_name, height=height, width=width)
        img_label_path = os.path.join(label_path, img_name.split('.')[0] + '.txt')
        df = pd.read_csv(
            img_label_path, 
            sep=' ', 
            names=['x1', 'y1', 'x2', 'y2', 'x3', 'y3', 'x4', 'y4', 'class_name', 'difficult']
        )
        df['xmin'] = df[['x1', 'x2', 'x3', 'x4']].min(axis=1)
        df['xmax'] = df[['x1', 'x2', 'x3', 'x4']].max(axis=1)
        df['ymin'] = df[['y1', 'y2', 'y3', 'y4']].min(axis=1)
        df['ymax'] = df[['y1', 'y2', 'y3', 'y4']].max(axis=1)
        df['width'] = df['xmax'] - df['xmin']
        df['height'] = df['ymax'] - df['ymin']
        df = df[['xmin', 'ymin', 'width', 'height', 'class_name']]

        for index, row in df.iterrows():
            category_name = row['class_name']
            try: 
                category_id = categories[category_name]
            except KeyError:
                categories[category_name] = category_id_counter
                category_id = category_id_counter
                category_id_counter += 1
                coco.add_category(CocoCategory(id=category_id, name=category_name))

            coco_image.add_annotation(
                CocoAnnotation(
                      bbox=[
                        row['xmin'], 
                        row['ymin'], 
                        row['width'], 
                        row['height']
                      ],
                      category_id=category_id,
                      category_name=category_name
                    )
                )

        coco.add_image(coco_image)
    coco_json = coco.json      
    save_json(coco_json, output_path)
    return categories, category_id_counter
```

```python
model_name = 'faster_rcnn'
model_dl_link = 'https://download.openmmlab.com/mmdetection/v2.0/faster_rcnn/faster_rcnn_r50_caffe_dc5_mstrain_1x_coco/faster_rcnn_r50_caffe_dc5_mstrain_1x_coco_20201028_233851-b33d21b9.pth'
config_file_name = 'faster_rcnn_r50_caffe_dc5_mstrain_1x_coco.py'
categories = load_json(categories_path)

# A Folder that will contain results
work_dir = os.path.join(extracted_data_root_dir, 'models' , model_name)
model_path = os.path.join(work_dir, 'model.pth')
config_path = os.path.join(work_dir, 'config.py')
```

```python
!mkdir -p $work_dir # Making a folder

config_path_temp = download_mmdet_config(
    model_name=model_name, 
    config_file_name=config_file_name
)

!mv $config_path_temp $config_path
!rm -rf mmdet_configs
!wget -O $model_path $model_dl_link
```

```python
train_pipeline = [
    dict(type='LoadImageFromFile'),
    dict(type='LoadAnnotations', with_bbox=True),
    dict(type='RandomCrop', crop_size=(height, width)),
    dict(
        type='Resize',
        img_scale=[(1333, 640), (1333, 672), (1333, 704), (1333, 736),
                   (1333, 768), (1333, 800)],
        multiscale_mode='value',
        keep_ratio=True),
    dict(type='RandomFlip', flip_ratio=0.5),
    dict(
        type='Normalize',
        mean=[103.53, 116.28, 123.675],
        std=[1.0, 1.0, 1.0],
        to_rgb=False),
    dict(type='Pad', size_divisor=32),
    dict(type='DefaultFormatBundle'),
    dict(type='Collect', keys=['img', 'gt_bboxes', 'gt_labels'])
]
```

```python
cfg = mmcv.Config.fromfile(config_path)

cfg.classes = list(categories.keys())

cfg.model.roi_head.bbox_head.num_classes = len(categories)

cfg.data_root = os.path.join(extracted_data_root_dir, 'data')

cfg.data.train.ann_file = train_anns_path
cfg.data.train.img_prefix = train_img_path
cfg.data.train.classes = cfg.classes 
cfg.data.train.pipeline = train_pipeline

# For sliced inference
cfg.data.val.ann_file = val_sliced_anns_path
cfg.data.val.img_prefix = val_sliced_img_path 
cfg.data.val.classes = cfg.classes 

# Without sliced inference
# cfg.data.val.ann_file = val_anns_path
# cfg.data.val.img_prefix = val_img_path
# cfg.data.val.classes = cfg.classes 

cfg.data.test.ann_file = test_sliced_anns_path
cfg.data.test.img_prefix = test_sliced_img_path
cfg.data.test.classes = cfg.classes 

cfg.log_config['hooks'] = cfg.log_config['hooks'] + [dict(type='TensorboardLoggerHook')]
cfg.work_dir = work_dir
cfg.runner = dict(type='EpochBasedRunner', max_epochs=10)
cfg.optimizer = dict(type='SGD', lr=1e-3, momentum=0.9, weight_decay=0.0001)
cfg.lr_config = dict(
    policy='CosineAnnealing',
    warmup='linear',
    warmup_iters=500,
    warmup_ratio=1e-2,
    min_lr=1e-5
)
cfg.load_from = model_path
cfg.dump(config_path)
```
